```python
test_db = parent_db.copy()
def check_for_error_div(div,month):
    check_val = list(test_db.iloc[:,month].value_counts())
    for i in check_val:
        if i > div:
            print("Error in Div")
            break;
    else :
        print("No Error")
        
div = 10

check_for_error_div(div,2)
```

```python
# include Annual as 13th
months_sorted_val = []
for m in range(1,14):
    val = list(test_db.iloc[:,m])
    val.sort()
    months_sorted_val.append(val)

def categorize_month_jan(value, month):
    case = months_sorted_val[month - 1].index(value)
    return (case//10)

for i in range(1,14):
    test_db[months[i] + '_CAT'] = test_db.iloc[:,i].apply(categorize_month_jan, args=(i,))
```

```python
category_db = test_db.iloc[:,14:]
category_db.head()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier 
from sklearn import metrics

x = category_db.iloc[:,:-1]
y = category_db.iloc[:,-1]
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=7)
print('x',x_train.shape, x_test.shape, 'Y', y_train.shape, y_test.shape)
```

```python
knn = KNeighborsClassifier(n_neighbors=3)
knn.fit(x_train,y_train)

y_predict = knn.predict(x_test)

print("knn Accuracy : ", metrics.accuracy_score(y_test, y_predict))
print("knn Accuracy : \n", metrics.confusion_matrix(y_test, y_predict))
```
